```python
import pandas as pd

def process_ingredients(data):
    final_list = []

    for idx, row in data.iterrows():
        cleaned_ingredients = row['Cleaned-Ingredients'].split(',')
        translated_ingredients = row['TranslatedIngredients'].split(',')

        for ingredient in cleaned_ingredients:
            ingredient = ingredient.strip().lower()  # Ensure consistent case and strip whitespace
            related_ingredients = [ing_amount.lower() for ing_amount in translated_ingredients if ingredient in ing_amount.lower()]

            for ing_amount in related_ingredients:
                final_list.append({
                    'TranslatedRecipeName': row['TranslatedRecipeName'],
                    'ingrident': ingredient,
                    'ingredient_relation': ing_amount
                })

    return pd.DataFrame(final_list)


processed_ingridients_df = process_ingredients(indian_food_dataset.iloc[:2])
processed_ingridients_df
```
